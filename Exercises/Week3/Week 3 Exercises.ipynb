```python
train_set_size = 2
test_set_size = 1000

# Train
x_train = np.random.rand(train_set_size, 1)
e_train = np.random.randn(train_set_size, 1)
y_train = x_train + e_train

# Test
x_test = np.random.rand(test_set_size, 1)
e_test = np.random.randn(test_set_size, 1)
y_test = x_test + e_test

# Create datasets
train_data = pr.gendatr(x_train, y_train)
test_data = pr.gendatr(x_test, y_test)

# Plot training data
pr.scatterr(train_data)

# reg_lambdas
reg_lambdas = [(float(i)*0.005) for i in range(10)]

for reg_lambda in reg_lambdas:
    # plot model
    w = pr.ridger(train_data, reg_lambda)
    pr.plotr(w)

    # Mean squares error
    mse = pr.testr(test_data, w)
    print(f"Lambda: {reg_lambda}, MSE: {mse}")
```
